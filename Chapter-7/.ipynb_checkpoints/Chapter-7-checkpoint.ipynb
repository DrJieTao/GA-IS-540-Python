```python
#Importing numpy and pandas

import numpy as np
import pandas as pd
pd.options.mode.chained_assignment= None
```

```python
#Creating a dataframe with random numbers of shape (5,4)
df= pd.DataFrame(np.arange(5*4).reshape(5,4))
df
```

```python
# Taking the random permutations of 3 
df_sampled_without_replacement = df.take(np.random.permutation(len(df))[:3])
df_sampled_without_replacement
```

```python
#generate a sample with replacement, we can draw random integers
sampler = np.random.randint(0, len(df), size=3)
sampler
```

```python
#These random integers can be used as input for the 'take' method, which is then used to sample the data.
df_sampled_with_replacement = df.take(sampler)
df_sampled_with_replacement
```

### Data Normalization or Standardization 

```python
df_norm = df / df.sum().astype('float')
df_norm
```

```python
perm = np.random.permutation(len(df))
perm
```

```python
df_permuted = df.take(perm)
df_permuted
```

```python
df_permuted_norm = df_permuted / df_permuted.sum().astype('float')
df_permuted_norm
```

### The Iris dataset example

```python
#Importing the data from the url given
fileURL = 'http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
#Reading the data usiing the URL 
data = pd.read_csv(fileURL, names=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width', 'Name'], header=None)
data.head()
```

```python
# Dropping the Nan values in the dataset
iris_data = data.dropna()
iris_data

```

```python
%matplotlib inline
my_colors = ['g', 'b','r','c']
#pl1 = pd.tools.plotting.scatter_matrix(iris_data, diagonal='hist', figsize=(10,10), color=my_colors, s=100)

# Plotting the graph using the scatter matrix graph
pl1 = pd.plotting.scatter_matrix(iris_data, diagonal='hist',figsize=(10,10),color=my_colors,s=100)
```

```python
%matplotlib inline

#Plotting the box plot to identify the outliers and to see how normalized the data is
pl2 = iris_data.boxplot(grid=False)
```

```python
# Normalizing the data , removing the outliers using the interpolate in Sepal Width
iris_data['Sepal Width'] = np.where((iris_data['Sepal Width'] > 4) | (iris_data['Sepal Width'] < 2.1), np.nan, iris_data['Sepal Width'])
iris_data['Sepal Width'] = iris_data['Sepal Width'].interpolate()
```

```python
#Plotting the boxplot 
pl3 = iris_data.boxplot(grid=False)
```

```python
# Taking random 100 instances from the dataset

sampler = np.random.randint(0, len(iris_data), size=100)
iris_data = iris_data.take(sampler)
iris_data.head()
```

```python
# Standardizing the data : using z-score normalization to rescale each feature. z-score normalization converts all features to a common scale with an average of zero and standard deviation of one
iris_data = iris_data.groupby('Name').transform(lambda x: (x-x.mean())/x.std(ddof=1))

```

```python
#Plotting the Boxplot 
pl4 = iris_data.boxplot(grid=False)

# We notice that each feature is normaized , showing an average zero and std 1 .
```

```python

# Plotting the Scatter matrix  of the standarized data.
pl5 = pd.plotting.scatter_matrix(iris_data, diagonal='hist', figsize=(10,10),color=my_colors, s=100)
```
