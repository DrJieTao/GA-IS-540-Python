```python
# Importing pandas for dataframes
import pandas as pd
```

```python
##RAW DATA
rawdata_users = pd.read_table('ml-1m/users.dat', sep='::', header=None)
rawdata_users[:5]
```

```python
# Creating a unames list for coloumn names in this dataframe
unames = ['user_id', 'gender', 'age', 'occupation', 'zip']

# Creating a dataframe "users" to read the table data from the source file and have the column names as the unames specified.
users = pd.read_table('ml-1m/users.dat', sep='::', header=None, names=unames)

#Checking for the 5 values using the index method.
users[:5]
```

```python
# Creating a rnames list for coloumn names in this dataframe

rnames = ['user_id', 'movie_id', 'rating', 'timestamp']

# Creating a dataframe "ratings" to read the table data from the source file and have the column names as the "rnames" specified.
ratings = pd.read_table('ml-1m/ratings.dat', sep='::', header=None, names=rnames)

#Checking for the 5 values using the index method.
ratings[:5]
```

```python
# Creating a mnames list for coloumn names in this dataframe
mnames = ['movie_id', 'title', 'genres']

# Creating a dataframe "movies" to read the table data from the source file and have the column names as the "mnames" specified.
movies = pd.read_table('ml-1m/movies.dat', sep='::', header=None, names=mnames)

#Checking for the 5 values using the index method.
movies[:5]
```

```python
# Merging the dataframes as a single dataframe "data"
# Merging the dataframes ratings and users first and then adding the movies to this table.
data = pd.merge(pd.merge(ratings, users), movies)

data[:15]

#Or you can have it data = pd.merge(ratings,users)
#data1 = pd.merge(data,movies)
#data1[:5], where the table matches the data in the table with their relavant movie ids

### Timestamp to be changed
```

```python
#Creating a Dataframe "mean_ratings" as a pivoted table based on the genders and also calculating the mean values of each gender for that movie 
mean_ratings = data.pivot_table('rating',index='title',columns='gender', aggfunc='mean')

#Checking the mean_ratings for the 5 movies
mean_ratings[:5]
```

```python
# Calculating the difference between the mean_ratings if the male and female genders fr each movie
#Adding the column "diff" to the existing mean_rating dataframe
mean_ratings['diff'] = mean_ratings['M'] - mean_ratings['F']

#Sorting the dataframe by creating a new dataframe "sorted_by_diff" , where the index of the sort is set to the difference.
sorted_by_diff = mean_ratings.sort_values(by='diff')

#checking the dataframe for an index value of 10 
sorted_by_diff[:10]

# Just in case we would need the absolute values for the difference
#abs(mean_ratings)
```

```python
# Sorting the dataframe "sorted_by_diff" from the bottom based on the lowest , where the men ratings are high 
sorted_by_diff[::-1][:10]
```
