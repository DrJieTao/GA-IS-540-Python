```python
# Importing the math, numpy , scipy and matplotlib modules for the construction of the basic-math graph
import math
import numpy as np
import scipy as sp 
import matplotlib.pyplot as plt
```

### Trying to plot the sin values using the numpy & matplotlib

```python
# Trying to plot the graph for the Sin(x) using the linspace of numpy
x = np.linspace(0,5)
plt.plot(x,np.sin(x))
plt.show()
```

```python
# Trying to plot the graph for the Cos(x) using the linspace of numpy
x= np.linspace(0,5)
plt.plot(x,np.cos(x))
plt.show()
```
